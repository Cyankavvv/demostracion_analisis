## Cálculo RMS y offset, para realizar corrección de offset

```python
def seccionar(v, i):
    indV0 = 1250*i
    indV400 = (1250 * i) + 300
    indV850 = (1250 * i) + 950
    indV1250 = (1250 * i) + 1250
    Vrms400 = v[indV0:indV400]
    Vrms850 = v[indV850:indV1250]
    VV1250 = v[indV0:indV1250]
    return Vrms400, Vrms850, VV1250
```

```python
def calcularRMS(Vrms400, Vrms850):
    Vrms400 = Vrms400**2
    RMS400 = sqrt((1/300) * np.sum(Vrms400))
    Vrms850 = Vrms850**2
    RMS850 = sqrt((1/300) * np.sum(Vrms850))

    return RMS400, RMS850
```

```python
def offsetIZQ(Vrms400, VV1250, vo, menosoff):
    if vo > 0:
        return np.array(VV1250) - vo*menosoff
    else:
        return np.array(VV1250) - vo*menosoff
```

```python
def offsetDER(Vrms850, VV1250, vo, menosoff):
    if vo > 0:
        return np.array(VV1250) - vo*menosoff
    else:
        return np.array(VV1250) - vo*menosoff
```

```python
def voltaje_corregido(v, numero):
    menosoff = np.array([1 for x in range(1250)])
    VV = []
    Vrms = []
    Voff = []
    for i in range(int(numero)):
        Vrms400, Vrms850, VV1250 = seccionar(v, i)

        RMS400, RMS850 = calcularRMS(Vrms400, Vrms850)
        if abs(RMS400) <= abs(RMS850):
            Vrms.append(RMS400)
            vo = sum(Vrms400)/len(Vrms400)
            Voff.append(vo)
            VV = VV + list(offsetIZQ(Vrms400, VV1250, vo, menosoff))
        else:
            Vrms.append(RMS850)
            vo = sum(Vrms850)/len(Vrms850)
            VV = VV + list(offsetIZQ(Vrms400, VV1250, vo, menosoff))
            Voff.append(vo)
    return VV, Vrms, Voff
```

```python
tiempo_inicial = time.time()
v, Vrms, Voff = voltaje_corregido(v, numero)
tiempo_final3 = time.time() - tiempo_inicial
```

Se vuelve a calcular el offset de los datos corregidos, para poder compararlos con los anteriores

```python
nuevoOffset = []
for i in range(int(numero)):
    indV0 = 1250*i
    indV400 = 1250 * i + 300
    indV850 = 1250 * i + 950
    indV1250 = 1250 * i + 1250
    
    Vrms400 = v[indV0:indV400]
    Vrms850 = v[indV850:indV1250]
    VV1250 = v[indV0:indV1250]
    
    voffn = sum(Vrms400)/len(Vrms400)
    voffn2 =  sum(Vrms850)/len(Vrms850)
    if voffn < voffn2:
        nuevoOffset.append(voffn)
    else:
        
        nuevoOffset.append(voffn2)
print(np.mean(nuevoOffset))
```

#### Se calculan las integrales

```python
inteac = []  #lista inteac contiene los valores de cada integral
for i in range(int(numero)):
    
    #Para indexar se utiliza xx e yy

    xx = 1250 * i + izquierda
    yy = (1250 * i) + derecha
     
    xss = list(t[xx:yy])
    yss = list(v[xx:yy])
    tiempo = xss[-1] - xss[0]
    #print(tiempo)
    
    inteac.append(sum(yss)*tiempo*-1/50)
#print(inteac)
```

```python
figure(figsize=(16, 9), dpi=80)

histo = plt.hist(inteac, bins = 106)
plt.ylabel('Absoulte frequency')
plt.xlabel('Charge')

media1 = [np.mean(inteac) for x in range(60)]
vay1 = [x for x in range(0, 60)]

plt.scatter(media1, vay1, c = 'red', marker = '|', label = f"Mean =  {np.mean(inteac)}")
plt.plot([], [], ' ', label=f"σ = {np.std(inteac)}")
plt.grid()
plt.legend()
plt.title("Integral's histogram")
#plt.savefig('HistogramaINTEGRAl.png', facecolor = 'lavender')
plt.show()
```
